```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def imshow(title,image,size=3):
    h,w = image.shape[:2]
    aspect_ratio = w/h
    plt.figure(figsize=(size*aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
im = cv2.imread('../images/images/opencv_inv.png',0)
imshow('Original',im)

kernel_val = np.ones((5,5),np.uint8)

ero = cv2.erode(im,kernel_val)
imshow('Erossion',ero)

dil = cv2.dilate(im,kernel_val)
imshow('Dilation',dil)

op = cv2.morphologyEx(im,cv2.MORPH_OPEN,kernel_val)
imshow('Open',op)

cl = cv2.morphologyEx(im,cv2.MORPH_CLOSE,kernel_val)
imshow('close',cl)
```

```python
i1 = cv2.imread('../images/images/londonxmas.jpeg')
imshow('Original',i1, size=7)
i2 = cv2.imread('../images/images/londonxmas.jpeg',0)
imshow('Grey',i2, size=7)

c1 = cv2.Canny(i2,50,120)
imshow('Canny1',c1,size=7)
```

```python
c2 = cv2.Canny(i2,10,100)
imshow('Canny Wide',c2,size=7)
c3 = cv2.Canny(i2,250,300)
imshow('Canny Narrow',c3,size=7)
```

```python
c4 = cv2.Canny(i2,60,120)
imshow('Canny 2',c4,size=7)
```

```python
def auto_canny(image):
    bl_im = cv2.blur(image,(5,5))
    med_val = np.median(image)
    up = int(min(0,0.66*med_val))
    dw = int(max(255,1.33*med_val))
    ed = cv2.Canny(bl_im,dw,up)
    return ed
```

```python
I = auto_canny(i2)
imshow("Auto Canny",I,size=7)
```
