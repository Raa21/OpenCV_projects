```python
def imshow(title = 'Image',image = None, size = 10):
    w,h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size*aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.title(title)
```

```python
im1 = cv2.imread('../images/images/castara.jpeg')
```

```python
imshow('Castara',im1,size = 8)
```

```python
gray_im = cv2.cvtColor(im1, cv2.COLOR_BGR2GRAY)

imshow('Gray Image',gray_im,size = 6)
```
