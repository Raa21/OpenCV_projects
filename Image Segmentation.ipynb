```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def imshow(title,image,size=5):
    h,w = image.shape[:2]
    aspect_ratio = w/h
    plt.figure(figsize=(size*aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
im = cv2.imread('../images/images/LP.jpg')
imshow("Original",im)
```

```python
im = cv2.imread('../images/images/LP.jpg')
grey = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
gauss = cv2.GaussianBlur(grey,(3,3),0)
imshow('Blur',gauss)
_,th = cv2.threshold(grey,0,255,cv2.THRESH_BINARY + cv2.THRESH_OTSU)
imshow('thresh',th)
count,heich = cv2.findContours(th,cv2.RETR_LIST,cv2.CHAIN_APPROX_NONE)
cv2.drawContours(im,count,-1,(0,255,0),2)
imshow('Cont',im)
print(len(count))
```

```python
im = cv2.imread('../images/images/LP.jpg')
grey = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
canny = cv2.Canny(grey,200,240)
imshow('Canny',canny)
count,heich = cv2.findContours(canny,cv2.RETR_LIST,cv2.CHAIN_APPROX_NONE)
cv2.drawContours(im,count,-1,(0,255,0),2)
imshow('Cont',im)
print(len(count))
```
