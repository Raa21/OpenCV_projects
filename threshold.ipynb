```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def imshow(title,image,size=7):
    w,h = image.shape[:2]
    aspect_ratio = w/h
    plt.figure(figsize=(size*aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
from skimage.filters import threshold_local
i1 = cv2.imread('../images/images/scan.jpg')
imshow('Image',i1)
v = cv2.split(cv2.cvtColor(i1,cv2.COLOR_BGR2HSV))[2]
T = threshold_local(v,25,offset=15,method='gaussian')

th = (v > T).astype('uint8')*255
imshow('Image',th,size=15)
```
