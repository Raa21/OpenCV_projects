```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def imshow(title = 'Image',image = None, size = 7):
    w,h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size*aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.title(title)
```

```python
im = cv2.imread('../images/images/Volleyball.jpeg')
imshow('Image',im)

h,w = im.shape[:2]
double_height,double_width = h/4, w/4
t = np.float64([[1,0,double_width],[0,1,double_height]])
t_im = cv2.warpAffine(im,t,(w,h))
imshow('Image2',t_im)
```

```python
i1 = cv2.imread('../images/images/oxfordlibrary.jpeg')
imshow('Image',i1)
```

```python
i_sc = cv2.resize(i1,None,fx=0.75,fy=0.75,interpolation=cv2.INTER_CUBIC)
imshow('sc1',i_sc)
i_sc1 = cv2.resize(i1,None,fx=2,fy=2,interpolation=cv2.INTER_LINEAR)
imshow('sc1',i_sc1)
i_sc2 = cv2.resize(i1,None,fx=2,fy=2,interpolation=cv2.INTER_AREA)
imshow('sc2',i_sc2)
i_sc3 = cv2.resize(i1,(900,600),interpolation=cv2.INTER_CUBIC)
imshow('sc3',i_sc3)
```

```python
h,w = i1.shape[:2]
st_r,st_cl = int(h*0.25),int(w*0.25)
en_r,en_cl = int(h*0.75),int(w*0.75)
c_i = i1[st_r:en_r,st_cl:en_cl]
imshow('Cropped',c_i)
c = i1.copy()
cv2.rectangle(c,(en_cl,en_r),(st_cl,st_r),(255,255,0),5)
imshow('Image',c)
```
